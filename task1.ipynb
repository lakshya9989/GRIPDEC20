# Importing all libraries required 

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
```

# Reading data from CSV file

```python
stu_info= pd.read_csv(r'C:\Users\91971\Desktop\vs\scores.csv')
stu_info.head(15)
```

# Plot the Data points on 2-D graph 

```python
stu_info.plot(x='Hours', y='Scores' ,style='r+')
plt.xlabel('No. of Hours Studied')  
plt.ylabel('Score(percentage)  ') 
plt.title('HOURS VS SCORES') 
plt.show()
```

# DATA PREPARTION 

```python
X = stu_info.iloc[:, :-1].values  
y = stu_info.iloc[:, 1].values 

```

# MODEL SELECTION

```python
# USE TRAIN_TEST_SPLIT MODEL IN SCIKIT-LEARN
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.25, random_state=0)
```

# TRAINING THE MODEL

```python
from sklearn.linear_model import LinearRegression 
reg = LinearRegression()  
reg.fit(X_train, y_train) 

# plotting the regression line 
slope=reg.coef_
intercept=reg.intercept_
line= slope*X+intercept   # similar to y=mx+C

plt.scatter(X,y,color = 'brown')
plt.plot(X,line)


```

# Evaluation / Prediction

```python
y_pred = reg.predict(X_test)
print(y_test)
print(y_pred)
```

### comparing the values

```python
tb= pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
print(tb)
```

### testing our own data

```python
Hours = 9.25
Hours=np.array(Hours).reshape(-1,1)     # convert hours into array as pred function takes array as input

new_pred = reg.predict(Hours)
print(f"No of Hours = {Hours}")
print(f"Predicted Score = {new_pred[0]}")
```

# EVALUATION OF MODEL 

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```
